```python
# b) Write a program to solve the 2 water jug problem by sequentially calling operators automatically.
def automatic_two_jug_problem(jug1_capacity, jug2_capacity, target):
    jug1, jug2 = 0, 0
    steps = []

    while jug1 != target and jug2 != target:

        if jug1 == 0:
            jug1 = jug1_capacity
            steps.append("Fill Jug1")

        elif jug2 == jug2_capacity:
            jug2 = 0
            steps.append("Empty Jug2")

        else:
            transfer = min(jug1, jug2_capacity - jug2)
            jug1 -= transfer
            jug2 += transfer
            steps.append("Transfer Jug1 -> Jug2")

        if jug1 == target or jug2 == target:
            break

    for step in steps:
        print(step)

    print(f"Success! Target of {target}L achieved.")

if __name__ == "__main__":
    automatic_two_jug_problem(3, 5, 4)
```

```python
""" B) Write a program to implement Missionaries and Cannibals problem. Constraints:
 three missionaries and three cannibals must cross a river using a boat which
  can carry at most two people, under the constraint that, for both banks,
   if there are missionaries present on the bank, they cannot be outnumbered by
   cannibals (if they were, the cannibals would eat the missionaries).
    The boat cannot cross the river by itself with no people on board. """



from collections import deque

def is_valid_state(m, c):

    if m < 0 or c < 0 or m > 3 or c > 3:
        return False
    if m > 0 and m < c:
        return False
    if (3 - m) > 0 and (3 - m) < (3 - c):

        return False
    return True

def get_next_states(state):
    m, c, b = state
    moves = [(1, 0), (0, 1), (1, 1), (2, 0), (0, 2)]
    next_states = []

    for move in moves:
        dm, dc = move
        if b == 1:
            new_state = (m - dm, c - dc, 0)
        else:
            new_state = (m + dm, c + dc, 1)

        if is_valid_state(new_state[0], new_state[1]):
            next_states.append(new_state)

    return next_states

def missionaries_and_cannibals():
    start_state = (3, 3, 1)
    goal_state = (0, 0, 0)

    queue = deque([(start_state, [])])
    visited = set()

    while queue:
        current_state, path = queue.popleft()

        if current_state in visited:
            continue
        visited.add(current_state)

        if current_state == goal_state:
            return path + [current_state]

        for next_state in get_next_states(current_state):
            queue.append((next_state, path + [current_state]))

    return None

def main():
    solution = missionaries_and_cannibals()
    if solution:
        print("Solution found!")
        for step in solution:
            print(step)
    else:
        print("No solution exists.")

if __name__ == "__main__":
    main()
```
